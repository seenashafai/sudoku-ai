## Part One
You should write all of your code for solving sudokus below this cell.

You must include a function called `sudoku_solver(sudoku)` which takes one sudoku puzzle (a 9x9 NumPy array) as input, and returns the solved sudoku as another 9x9 NumPy array. This is the function which will be tested. 

```python
import numpy as np
sudokus = np.load("/Users/seena/PycharmProjects/untitled/AI/hard_puzzle.npy")
sudoku = sudokus[9]

digits = '123456789'
rows = 'ABCDEFGHI'


def cross(A, B):
    # Returns cross-product of A and B
    return [a + b for a in A for b in B]

#Define units of each row, column and 3x3 box for searching
rowList = []
colList = []
boxList = []
for i in rows:
    rowList.append(cross(i, digits))
for i in digits:
    colList.append(cross(rows, i))
for i in ['ABC', 'DEF', 'GHI']:
    for j in ['123', '456', '789']:
        boxList.append(cross(i, j))

#UNIT = Row, Column and Box for a given square
unitList = rowList + colList + boxList  # Combine arrays for each row, column and box


squares = cross(rows, digits)
#Using list comprehension to improve readability-----------
#Create dictionary of all possible units for rapid searching
unitDict = dict((box, [unit for unit in unitList if box in unit]) for box in squares)
#Create dictionary of all possible peers for rapid searching using set subtraction to remove duplicates
peerDict = dict((box, set(sum(unitDict[box], [])) - set([box])) for box in squares)

#Verify that the sudoku is valid
def valid(sudoku, end): #-> Boolean
    def rowChecker(s):
        #Check array for consecutive nonzero digits
        matrix = s
        for i in range(0, len(s)):
            for j in range(0, len(s)-1):
                if s[i][j] == s[i][j+1] and s[i][j] != 0:
                    #Return False (invalid) if consecutive nonzero values detected
                    return False
                if end:
                    #Extra check before solved sudoku is output
                    if len(matrix[i]) != len(set(matrix[i])): #Checks for duplicates
                        return False
        return True

    #Transpose to convert rows to columns, and check again using same method
    transposed = [[sudoku[j][i] for j in range(len(sudoku))] for i in range(len(sudoku[0]))]
    #Validate rows and columns
    if rowChecker(sudoku) and rowChecker(transposed):
        return True
    return False


# Converts grid into dictionary of possible values
def sudoku_solver(sudoku):
    sudokuDict = {}
    if valid(sudoku, False):
        for i, row in enumerate(sudoku):  # i = row number
            for j, val in enumerate(row):  # j = column number
                # Define square ID (A1...I9)
                square = rows[i] + digits[j]
                # If empty space detected
                if val == 0:
                    # Assign possible values 1-9 to dictionary
                    sudokuDict[square] = '123456789'
                else:
                    # Otherwise, assign provided value to dictionary
                    sudokuDict[square] = str(sudoku[i][j])
        #Pass dictionary to solve function
        result = solve(sudokuDict)
        #Handle returned tuple: (Bool, Dictionary)
        if result[0]:
            return output(result[1])
        elif not result[0]:
            return (np.ones((9,9)) * -1)
    else:
        return (np.ones((9,9)) * -1)


def eliminate(sDict):
    # iterate through grid dictionary
    for i, (key, val) in enumerate(sDict.items()):
        # If square's dictionary contains more than one item
        if len(val) > 1:
            peers = peerDict[key] #Get all intersecting cells using square's key (e.g. A1)
            peer_values = []
            #Get the values of the intersecting cells
            for i in peers:
                if len(sDict[i]) == 1:
                    peer_values.append(sDict[i])
            #Convert to set to allow for set subtraction and automatic duplication removal
            peer_values = set(peer_values)
            #Eliminate values already in the peer list
            sDict[key] = ''.join(set(sDict[key]) - peer_values)
    return sDict


def naked_single(sDict):
    # Iterate through each unit
    for i in unitList:
        for j in digits:
            # List comprehension: adds every occurance of each digit in a unit into an array
            apps = [square for square in i if j in sDict[square]]
            # Sets to sole dictionary value if only appears once in unit
            if len(apps) == 1:
                sDict[apps[0]] = j
    return sDict


def apply(sDict):
    stuck = False
    while not stuck:
        #Check number of values in dictionary before solving
        arr = []
        for i in enumerate(sDict.keys()):
            if len(i) == 1:
                arr.append(i)
        before = len(arr)

        # Apply solving strategies
        sDict = eliminate(sDict)
        sDict = naked_single(sDict)

        #Check number of values in dictionary after solving
        arr = []
        for i in enumerate(sDict.keys()):
            if len(i) == 1:
                arr.append(i)
        after = len(arr)

        # Program is stuck if no values have changed
        stuck = before == after

        #Check if any squares have 0 candidates
        if len([box for box in sDict.keys() if len(sDict[box]) == 0]):
            return False #currently unsolveable
    return sDict


def solve(sDict):
    sDict = apply(sDict)
    if sDict == False:
        return (False, sDict)

    # If all squares contain a single value, return solved dictionary
    if all(len(v) == 1 for k,v in sDict.items()):
        return (True, sDict)

    # Identify square with least number of possibilities
    length, k = min((len(val), key) for key, val in sDict.items() if len(val) > 1)

    for i in sDict[k]:
        #Recursively set up new boards with alternate configurations
        newBoard = dict(list(sDict.items()))
        #add new config to board
        newBoard[k] = i
        #Attempt to solve new config
        re_solve = solve(newBoard)
        try:
            #Return solved board
            if re_solve[0]:
                return re_solve
        except:
            return (False, sDict)
    return (False, sDict)



def output(sDict):
    outArray = []
    #Guard statement: Check that all squares contain a single candidate
    if all(len(v) == 1 for k,v in sDict.items()):
        #Formatting: Convert dictionary back into NumPy array
        for i in sDict:
            outArray.append(int(sDict[i]))

        #Split array of dictionary values into rows
        slices = [outArray[x:x + 9] for x in range(0, len(outArray), 9)]
        #Convert rows into 2D array and return
        outarr = np.asarray(slices)
        if valid(outarr, True): #Final validation check
            return np.asarray(slices)
        else:
            return (np.ones((9,9)) * -1)


```
